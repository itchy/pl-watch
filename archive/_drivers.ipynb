```python
#https://beautiful-soup-4.readthedocs.io/en/latest/#kinds-of-objects

import requests, json
from bs4 import BeautifulSoup
import re


def get_car_number(first_name, last_name):
    first_name = first_name.lower()
    last_name = last_name.lower()
    # Handle known drivers with fixed numbers
    # Don't use temp number 1 for WDC
    if first_name == "lando" and last_name == "norris":
        return "4"
    if first_name == "max" and last_name  == "verstappen":
        return "3"
    if first_name == "lewis" and last_name == "hamilton":
        return "44"
    if first_name == "charles" and last_name == "leclerc":
        return "16"
    if first_name == "george" and last_name == "russell":
        return "63"
    if first_name == "fernando" and last_name == "alonso":
        return "14"
    if first_name == "sergio" and last_name == "perez":
        return "11"
    if first_name == "carlos" and last_name == "sainz":
        return "55"
    if first_name == "valtteri" and last_name == "bottas":
        return "77"
    if first_name == "pierre" and last_name == "gasly":
        return "10"
    if first_name == "lance" and last_name == "stroll":
        return "18"
    if first_name == "kevin" and last_name == "magnussen":
        return "20"
    if first_name == "yuki" and last_name == "tsunoda":
        return "22"
    if first_name == "mick" and last_name == "schumacher":
        return "47"
    if first_name == "nicholas" and last_name == "latifi":
        return "6"
    if first_name == "sebastian" and last_name == "vettel":
        return "5"
    if first_name == "zhou" and last_name == "guanyu":
        return "24"
    if first_name == "alexander" and last_name == "albon":
        return "23"
    if first_name == "logan" and last_name == "sargent":
        return "2"
    # Send an HTTP request to the website
    url = "https://www.formula1.com/en/drivers/{}".format(first_name + "-" + last_name)
    response = requests.get(url)

    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the title of the page
    # title = soup.title.string
    number_p = soup.find_all('p', class_='typography-module_body-xs-semibold__Fyfwn typography-module_lg_body-s-compact-semibold__cpAmk')[2]
    if number_p:
        car_number = number_p.text.strip()
        return car_number
    return "00"

def get_rows(year):
        # Send an HTTP request to the website
    url = "https://www.formula1.com/en/results/{}/drivers".format(year)
    response = requests.get(url)

    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the title of the page
    title = soup.title.string
    print(title)
    # print(response.content)
    rows = soup.find_all('tr', class_='Table-module_body-row__shKd-')
    print(f"Found {len(rows)} drivers")

    if len(rows) == 1:
        return get_rows(year-1)
    return rows

def get_drivers(year):
    rows = get_rows(year)
    drivers = []
    for r in rows:
        elements = r.find_all('td')
        spans = elements[1].find_all("span", class_='max-lg:hidden')

        first_name = elements[1].find_all("span", class_='max-lg:hidden')[0].text
        last_name  = elements[1].find_all("span", class_='max-md:hidden')[0].text
        place = elements[0].text
        points = elements[-1].text
        d = {}
        d['first_name'] = first_name    
        d['last_name'] = last_name
        d['place'] = place
        d['points'] = points
        d['car_number'] = get_car_number(first_name, last_name)
        drivers.append(d)

    return drivers

year = 2026

# Serialize the data to a JSON formatted string
json_string = json.dumps(get_drivers(year), indent=4)

# Write the JSON string to a file
with open(f"{year}_drivers.json", "w") as outfile:
    outfile.write(json_string)
```
