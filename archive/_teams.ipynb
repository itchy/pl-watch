```python
#https://beautiful-soup-4.readthedocs.io/en/latest/#kinds-of-objects

import requests, json
from bs4 import BeautifulSoup


def get_rows(year):
    # Send an HTTP request to the website
    url = "https://www.formula1.com/en/results/{}/team".format(year)
    response = requests.get(url)

    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the title of the page
    title = soup.title.string
    print(title)

    # Find all the tables on the page
    table = soup.find('tbody', class_='Table-module_tbody__KEiSx')
    rows = table.find_all('tr')
    print(f"Found {len(rows)} teams")
    if len(rows) == 1:
        return get_rows(year-1)
    return rows

def get_teams(year):
    rows = get_rows(year)
    teams = []
    for r in rows:
        elements = r.find_all('td')
        team_name = elements[1].find('a').text
        place = elements[0].text
        points = elements[-1].text
        t = {}
        t['team_name'] = team_name
        t['place'] = place
        t['points'] = points
        teams.append(t)
    return teams

year = 2026

# Serialize the data to a JSON formatted string
json_string = json.dumps(get_teams(year), indent=4)

# Write the JSON string to a file
with open(f"{year}_teams.json", "w") as outfile:
    outfile.write(json_string)
```
